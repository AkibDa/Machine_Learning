```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv("data/income.csv")
df
```

```python
%matplotlib inline
plt.xlabel("Year")
plt.ylabel("Income")
plt.scatter(df.year, df[['per capita income (US$)']], color='red')
```

```python
regression = LinearRegression()
regression.fit(df[['year']], df[['per capita income (US$)']] )
```

```python
prediction = regression.predict(pd.DataFrame({'year': [2020]}))
print(f"Predicted value for 2020: {prediction[0][0]:,.2f}")
```
