```python
import pandas as pd
d = pd.read_csv('data/carprices.csv')
d
```

```python
dummies = pd.get_dummies(d['Car Model'])
dummies
```

```python
merged = pd.concat([d, dummies], axis=1)
merged
```

```python
final = merged.drop(["Car Model",2],axis='columns')
final
```

```python
X = final.drop('Sell Price($)',axis='columns')
X
```

```python
y = final['Sell Price($)']
y
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
model.fit(X, y)
```

```python
model.score(X,y)
```

```python
model.predict([[45000,4,0,0]])
```

```python
model.predict([[86000,7,0,1]])
```
